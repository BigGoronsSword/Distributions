```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics
%matplotlib inline

bernoulli = np.random.binomial(1, .5, 100)
plt.hist(bernoulli, color = 'purple')
print(bernoulli.mean())
print(statistics.stdev(bernoulli))
plt.axvline(bernoulli.mean(), color='g', linestyle='solid', linewidth=2)

plt.axvline(bernoulli.mean() + bernoulli.std(), color='g', linestyle='dashed', linewidth=1)

plt.axvline(bernoulli.mean()-bernoulli.std(), color='g', linestyle='dashed', linewidth=1) 
plt.show
#The mean and std do not articulate this data well. There are no data points that are close to the mean, and the standard
#deviation lines fall on the bins. The graph alone better describes the data as the mean could be construed as misleading.
```

```python
import numpy as np
import statistics
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
binomial = np.random.binomial(30, .5, 100)
plt.axvline(binomial.mean(), color='purple', linestyle='solid', linewidth=2)

plt.axvline(binomial.mean() + binomial.std(), color='purple', linestyle='dashed', linewidth=1)

plt.axvline(binomial.mean()-binomial.std(), color='purple', linestyle='dashed', linewidth=1) 
plt.hist(binomial, color = "green")
print(binomial.mean())
print(statistics.stdev(binomial))
#The mean and std better describe this graph. Together, assuming normality, the mean and std show that 95% of data fall between 
#10 and 20, which looks about right. The graph is not a pretty bell curve but the numbers decently describe the spread.
```

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1. 
Add them together to create a third variable.
Graph the third variable using a histogram.
Compute the mean and standard deviation and plot them as vertical lines on the histogram.
Evaluate the descriptive statistics against the data.

```python
import numpy as np
import statistics
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
var1 = np.random.normal(5, .05, 100)
var2 = np.random.normal(10, 1, 100)

var3 = var1 + var2

plt.axvline(var3.mean(), color='blue', linestyle='solid', linewidth=2)

plt.axvline(var3.mean() + var3.std(), color='blue', linestyle='dashed', linewidth=1)

plt.axvline(var3.mean()-var3.std(), color='blue', linestyle='dashed', linewidth=1) 
plt.hist(var3, color = "purple")
print(var3.mean())
print(statistics.stdev(var3))
#Most of these distributions, with the exception of normal, appear to not be truly reflected by the mean and standard
#deviation.
```
